```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
dice = np.array(range(1, 7))

def dice_throw(dice):
    # simulate a simple dice throw
    return np.random.choice(dice)

avgs = []
n_throws = []

number_of_throws = [5, 10, 15, 20, 25, 50, 75, 100, 125, 150, 175, 200, 400, 500, 600, 700, 800, 900, 1000, 2000, 3000, 4000]
throws = []

for i in number_of_throws:
    throws = []

    # here we simulate for every throw
    for j in range(i):
        throws.append(dice_throw(dice))

    throws.append(dice_throw(dice))

    mean = np.mean(throws)
    avgs.append(mean)
    n_throws.append(i)

print('Number of throws', n_throws)
print('Averages', avgs)
```

```python
# Plot throws
plt.xkcd()
plt.figure(figsize=(8,5))
plt.plot(n_throws, avgs, color='xkcd:brown', zorder = 5)
plt.scatter(n_throws, avgs, alpha=0.5, s=100, edgecolor='black')
plt.axhline(y = 3.5, linestyle='--', lw=3, zorder=1,color='xkcd:terra cotta')
plt.xticks(fontsize=15)
plt.yticks(fontsize=15)
plt.gca().spines['top'].set_visible(False)
plt.gca().spines['right'].set_visible(False)
plt.xlabel("Number of dice throw", fontsize=15)
plt.ylabel("Running average", fontsize=15)

# Annotate
plt.annotate(f'Average across all throws', (number_of_throws[-1], 3.55), fontsize=12, color='xkcd:terra cotta')
plt.title("Running average of dice throw simulation",fontsize=14)
plt.show()
```
